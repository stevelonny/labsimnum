```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']
states = ["ground", "first"]
steps = ["unif", "gauss"]
real_r = [1.5, 5.0]
```

```python
for i, stase in enumerate(["ground", "first"]):
    plt.figure(figsize=(16,5))
    for j, step in enumerate(["unif", "gauss"]):
        n_throws, cum_avg, block_error = np.loadtxt("../data/ex_05_{state}_{step}.dat".format(state=stase, step=step), unpack=True, usecols=(0,1,2))
        colour = colors[i]
        plt.subplot(1,2,j+1)
        plt.errorbar(n_throws, cum_avg-real_r[i], yerr=block_error, color=colour)
        plt.axhline(y=0, color='r', linestyle='-.')
        plt.title("Average position - {state} state, {step} step".format(state=stase, step=step))
        plt.xlabel('Number of blocks')
        plt.ylabel(r"$\langle r \rangle / a_0$")
        plt.grid(True)

        # plt.subplot(1,2,2)
        # plt.plot(n_throws, block_error, color = colour)
        # plt.title("Block error")
        # plt.xlabel('Number of blocks')
        # plt.ylabel('Block error')

    plt.show()
```
